**XML:**

Filtre los discos para que muestre aquellos que esten entre el año 1990 y 1997

Muestre con la función iloc cual es el disco más viejo y el más reciente.


```python
df_xml[(df_xml['YEAR']>=1990) & (df_xml['YEAR']<= 1997)]
```

```python
#Funcion iloc

#iloc[0] -> mas antiguo
#iloc[-1] -> mas reciente

df_ordenado=df_xml.sort_values('YEAR')

disco_mas_viejo = df_ordenado.iloc[0]
disco_mas_reciente= df_ordenado.iloc[-1]

print("*** Disco mas viejo: ***\n", disco_mas_viejo)
print("\n")
print("*** Disco mas reciente: ***\n", disco_mas_reciente)
```

```python
df_xls.head()

# Quitar columnas que no se ocuparan: Id de Empleado, Id de Producto, Id de Cliente, Id de Pedido, Fecha de pedido

```

```python
#Borrar columnas
df_xls=df_xls.drop(columns=['Id de empleado','Id de producto','Id de cliente','Id de pedido','Fecha de pedido'])
```

```python
#Copia de dataframe con datos necesarios
df_xls_bk =df_xls.copy()
```

```python
df_xls_bk.head() #Verificando que solo tenga las columnas necesarias
```

```python
df_xls.groupby('Empleado')['Ventas'].sum() #Total de ventas por Vendedor
```

```python
#Categoria que mas vende y que menos vende

ventas_categorias=df_xls.groupby('Categoría')['Ventas'].sum() #Sumar ventas por categorias

print("Categoria que mas vende: ",ventas_categorias.idxmax()) #la que mas vende
print("\n")
print("Categoria que menos vende: ", ventas_categorias.idxmin()) #la que menos vende
```
